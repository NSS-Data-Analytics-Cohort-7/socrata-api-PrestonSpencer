### 4) Make an API request that returns the entire Top 500 Monthly Searches dataset. Make a chart that shows the number of times "maps" was searched in a month across the entire time frame.

```python
full_df = pd.DataFrame(res, columns = ['month_name', 'year', 'query_count', 'query_text'])
full_df.head()
```

```python
maps_df = full_df[full_df['query_text'].eq('maps')]
maps_df
```

```python
maps_df['Month'] = maps_df['month_name'] + ' ' + maps_df['year']
maps_df.rename(columns = {'query_count':'Query Count'}, inplace=True)
maps_df
```

```python
maps_df.info()
```

```python
maps_df['Query Count'] = maps_df['Query Count'].astype(str).astype(int)
maps_df.info()
```

```python
plt.figure(figsize = (5, 5))
plt.bar('Month', 'Query Count', data = maps_df)
plt.ylim([0, 150])
plt.show();
```

```python
maps_df_2 = full_df[full_df['query_text'].str.contains('maps')]
maps_df_2
```

```python
maps_df_2['Month'] = maps_df_2['month_name'] + ' ' + maps_df_2['year']
maps_df_2
```

```python
maps_df_2['query_count'] = maps_df_2['query_count'].astype(str).astype(int)
maps_df_2.info()
```

```python
plt.figure(figsize = (5, 5))
sns.scatterplot('Month', 'query_count', hue = 'query_text', data = maps_df_2)
plt.ylim([0, 150])
plt.show();
```
